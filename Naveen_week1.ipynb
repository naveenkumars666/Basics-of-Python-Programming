```python
#print even and odd characters in a given string 

n=str(input())
odd=""
even=""
for i in range(0,len(n)):
   if i%2==0:
     even=even+n[i]
   else:
     odd=odd+n[i]
print("even characters:",even)
print("odd characters:",odd)
```
